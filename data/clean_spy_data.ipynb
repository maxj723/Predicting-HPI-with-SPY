```python
import pandas as pd
```

```python
data = pd.read_csv('spy.csv')
```

```python
# convert data to monthly data
data['Date'] = pd.to_datetime(data['Date'])
data = data.set_index('Date')
data = data['Close'].resample('ME').mean()
data
```

```python
# convert date index to monthly
data.index = data.index.to_period('M')
data
```

```python
len(data)
```

```python
# save new csv
data.to_csv('spy_monthly.csv')
```
